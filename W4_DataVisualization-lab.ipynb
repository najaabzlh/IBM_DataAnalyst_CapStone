# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

#Get the master table data to dataframe for easy processing
import matplotlib.pyplot as plt
query = """
        SELECT * FROM master
        """

df2 = pd.read_sql_query(query,conn)
df2.drop('index',axis=1,inplace=True)

#Plot histogram of ConvertedComp
df2['ConvertedComp'].plot(kind='hist')
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

df2['Age'].plot(kind='box')
plt.title('Box plot of Age')
plt.ylabel('Frequency')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

df2.plot(kind='scatter', x='Age',y='WorkWeekHrs', color='darkblue')
plt.title('WorkWeekHrs by Age')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
s = df2['Age'] * 10    # Create a new array for s

# Create a scatter plot
plt.scatter(df2['WorkWeekHrs'], df2['CodeRevHrs'], s=s, alpha=0.5)

# Add labels and a title
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs')

# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# get data on database desire to learn next year
query_db = """
        SELECT DatabaseDesireNextYear,SUM(Respondent) as 'TotalRespondentChoose'
        FROM DatabaseDesireNextYear
        GROUP By DatabaseDesireNextYear
        ORDER By SUM(Respondent) DESC
        LIMIT 5
"""

df_dblearn = pd.read_sql(query_db,conn)
df_dblearn

#create pie chart
df_dblearn['TotalRespondentChoose'].plot(kind='pie',
                                figsize=(15, 6),
                                autopct='%1.1f%%',
                                startangle=90,
                                shadow=False,
                                labels=df_dblearn['DatabaseDesireNextYear']
                                )

plt.title('Top 5 Databases Respondent Wish To Learn Next Year', y=1.12)
plt.axis('equal')

plt.legend(labels=df_dblearn['DatabaseDesireNextYear'], loc='upper left')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Filter data for ages between 30 and 35
df_filtered_aged = df2[(df2['Age'] >= 30) & (df2['Age'] <= 35)]

# Group data by age and calculate the median of WorkWeekHrs and CodeRevHrs
df_grouped_age = df_filtered_aged.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create a stacked chart
df_grouped_age.plot(kind='bar', stacked=True)

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Ages 30-35')

# Show the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_filtered_age= df2[(df2['Age'] >= 45) & df2['Age'] <= 60]

# Group data by age and calculate the median of ConvertedComp
df_grouped = df_filtered_age.groupby('Age')['ConvertedComp'].median()

# Create a line chart
plt.plot(df_grouped.index, df_grouped.values)

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45-60')

# Show the plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
df_grouped = df2.groupby('MainBranch').size()
# Create a horizontal bar chart
df_grouped.plot.barh()

# Add labels and a title
plt.xlabel('Frequency')
plt.ylabel('Main Branch Category')
plt.title('Horizontal Bar Chart of Main Branch Categories')

# Show the plot
plt.show()
```
